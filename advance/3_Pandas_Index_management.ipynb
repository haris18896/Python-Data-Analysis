# Index Management

1. df.reset_index()
2. df.set_index()
3. df.sort_index()

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.index
```

```python
df.index.name = 'job_index'
df.head()
```

# reset_index()

```python
df_usa = df[df['job_country'] == 'United States']
```

```python
df_usa
```

```python
df_usa.reset_index(inplace=True)
df_usa
```

# set_index()

```python
df_usa.set_index('index', inplace=True)
df_usa
```

# sort_index()

```python
median_pivot = df_usa.pivot_table(index='job_title_short', values='salary_year_avg', aggfunc=['median', 'min', 'max'])
median_pivot
```

```python
median_pivot.sort_values(by=('median', 'salary_year_avg'), ascending=True)
median_pivot
```

```python
median_pivot.sort_index(inplace=True)
```

```python
median_pivot
```

# Exercise: Job Demand

```python
df_US = df[df['job_country'] == 'United States']
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
df_US['job_posted_month']
```

```python
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_US_pivot
```

```python
df_US_pivot.reset_index(inplace=True)
df_US_pivot['month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month
```

```python
df_US_pivot
```

```python
df_US_pivot.sort_values('month_no', inplace=True)
df_US_pivot.set_index('job_posted_month', inplace=True)
df_US_pivot.drop(columns='month_no', inplace=True)
df_US_pivot
```

```python
df_US_pivot.plot(kind='line')
```

```python
top_3 = df_US['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()
df_US_pivot[top_3].plot(kind='line')
```
