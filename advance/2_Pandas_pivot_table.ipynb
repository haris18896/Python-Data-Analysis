# Pandas : Pivot Table

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.pivot_table(index='job_posted_date', aggfunc='size').plot()
```

```python
df.pivot_table(index='job_title_short', values='salary_year_avg', aggfunc='median')
```

```python
df.pivot_table(index='job_title_short', values='salary_year_avg', aggfunc='median').plot()
```

```python
df.pivot_table(index='job_title_short', values='salary_year_avg', columns='job_country', aggfunc='median')
```

```python
top_countries = df['job_country'].value_counts().head(6).index
df_job_country_salary = df.pivot_table(index='job_country', values='salary_year_avg', columns='job_title_short', aggfunc='median')

df_job_country_salary = df_job_country_salary.loc[top_countries]

job_titles = ['Data Analyst', 'Data Scientist', 'Data Engineer']

df_job_country_salary = df_job_country_salary[job_titles]

df_job_country_salary.plot(kind='bar')
plt.ylabel('Median Salary ($USD)')
plt.title('Median Salary in Country and Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()
```
