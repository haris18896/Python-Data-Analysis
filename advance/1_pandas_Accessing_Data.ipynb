````python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# loading Data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
````

````python
df.loc[:, 'salary_rate':'salary_hour_avg'].dropna(subset='salary_rate')
````

# Pandas : Clean UP

````python
median_salary_avg = df['salary_year_avg'].median()
````

````python
median_salary_hour = df['salary_hour_avg'].median()
````

````python
df_filled = df
df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_salary_avg)
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_salary_avg)
````

### Drop Duplicate

````python
df_unique = df_filled.drop_duplicates()
df_unique
````

````python
df_unique = df_unique.drop_duplicates(['job_title', 'company_name'])
df_unique
````
