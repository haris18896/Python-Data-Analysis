# Pandas: Merging Data Frames

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")
df  = dataset['train'].to_pandas()

# Data Cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

### First Data Frame

```python
df_US_software_pivot = pd.read_csv('https://lukeb.co/software_csv', index_col='job_posted_month')
df_US_software_pivot
```

### Second Data Frame

```python
df_US = df[df['job_country'] == 'United States']
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_US_pivot.reset_index(inplace=True)
df_US_pivot['month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month
df_US_pivot.sort_values('month_no', inplace=True)
df_US_pivot.set_index('job_posted_month', inplace=True)
df_US_pivot.drop(columns='month_no', inplace=True)
df_US_pivot
```

```python
df_US_pivot.merge(df_US_software_pivot, on='job_posted_month', how='inner')
```

```python
df_US_pivot.merge(df_US_software_pivot, on='job_posted_month', how='outer')
```

```python
df_US_merged = df_US_pivot.merge(df_US_software_pivot, on='job_posted_month', how='inner')
top_5 = df_US_merged.sum().sort_values(ascending=False).head(5).index.to_list()
top_5
```

* With the help of Parenthesis we can break the code into multiple lines

```python
top_5 = (
    df_US_merged
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .index
    .to_list()
)
top_5
```

```python
df_US_merged[top_5].plot(kind='bar', figsize=(10, 5))
```

```python
df_US_merged[top_5].plot(kind='line', figsize=(10, 5))
```
